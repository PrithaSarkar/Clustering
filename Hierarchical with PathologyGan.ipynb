```python
import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, v_measure_score
```

```python
pathologygan = h5py.File('pge_dim_reduced_feature.h5', mode='r')
list(pathologygan.keys())
```

```python
dataset_fname=pathologygan['file_name']
dataset_pca=pathologygan['pca_feature']
dataset_umap=pathologygan['umap_feature']
```

```python
ground_truth=dataset_fname[:]
```

```python
X_pca=dataset_pca[:]
```

```python
dendrogram_pca_ward=sch.dendrogram(sch.linkage(X_pca, method='ward'))
```

```python
hc_pca_ward=AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
```

```python
y_hc_pca_ward=hc_pca_ward.fit_predict(X_pca)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(X_pca[y_hc_pca_ward == 0, 0], X_pca[y_hc_pca_ward == 0, 1], color = '#93d3f7', label = 'Cluster 1',s=50)
plt.scatter(X_pca[y_hc_pca_ward == 1, 0], X_pca[y_hc_pca_ward == 1, 1], color = '#c793f7', label = 'Cluster 2',s=50)
plt.grid(False)
plt.title('Clusters of Tissues')
plt.legend()
plt.show()
```

```python
score = silhouette_score(X_pca, y_hc_pca_ward, metric='euclidean')
print('Silhouette Score is {} when number of clusters are {}'.format(score, 2))
vms=v_measure_score(ground_truth, y_hc_pca_ward)
print('V Measure Score is {} when number of clusters are {}'.format(vms, 2))
```

```python
dendrogram_pca_centroid=sch.dendrogram(sch.linkage(X_pca, method='centroid'))
```
